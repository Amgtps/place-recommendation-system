```python
import pandas as pd
import json
import os 
import codecs


from tqdm import tqdm 
```

```python
with codecs.open('timeandmonth.json', 'r',encoding='utf-8') as f: 
    file = json.load(f)
    
file[0]
```

```python
dfs = []

for place in file:

    #n = len(place['name']) 

    df = {
        'state_name':[place['state_name']],
        'city_name':[place['city_name']], 
        #'name':place['name'], 
        # 'description':place['description'],
        'best_time':[place['best_time']],
        'ideal_duration':[place['ideal_duration']],
        

    }
    print(df) 

    df = pd.DataFrame(df) 
    dfs.append(df) 

    
    
```

```python
pd.concat(dfs).to_csv('new_city.csv', index = False ) 
```
